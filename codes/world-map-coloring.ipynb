```python
import geopandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries_continents = pd.read_csv('./countries_communities.csv')
countries_continents.set_index('country', inplace=True)
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world
```

```python
world = world.join(countries_continents, on='iso_a3')
```

```python
plt.figure(figsize=(15, 12))
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().set_xticks([])
plt.gca().set_yticks([])
world.plot(column='community', ax = plt.gca(), missing_kwds={'color': 'lightgrey'})
```
