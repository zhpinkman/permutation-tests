```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
observed = 133599.4
```

```python
with open('./temp.txt', 'r') as f:
    a = f.read().splitlines()
    f.close()
```

```python
a = [int(float(i)) for i in a]
```

```python
with open('./file.txt', 'r') as f:
    b = f.read().splitlines()
    f.close()
```

```python
b = [int(float(i)) for i in b]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(a, color='salmon', label='pre-network permutation networks')
plt.hist(b, color='skyblue', label='node-permutation networks')
plt.vlines(x=observed, colors='black', ymin=0, ymax=250, label='observed network')
plt.title('Comparison between the networks generated by pre-network permutation and node-permutation networks')
plt.xlabel('Statistic for the effect in interest')
plt.ylabel('Frequency')
plt.legend(loc='upper left')
```
