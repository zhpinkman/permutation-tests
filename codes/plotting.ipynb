```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib
```

```python
observed = 133599.4
```

```python
with open('./ergm-merchandise.txt', 'r') as f:
    a = f.read().splitlines()
    f.close()
```

```python
a = [int(float(i)) for i in a]
```

```python
with open('./general-merchandise.txt', 'r') as f:
    b = f.read().splitlines()
    f.close()
```

```python
b = [int(float(i)) for i in b]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(a, color='gold', label='Pre-network Permutation Networks')
plt.hist(b, color='gainsboro', label='Node-permutation Networks')
plt.vlines(x=observed, colors='black', ymin=0, ymax=250, label='Observed Network')
plt.title('Comparison between the Networks Generated by Pre-network Permutation and Node-permutation Networks')
plt.xlabel('Statistic for the Effect in Interest')
plt.ylabel('Frequency')
plt.legend(loc='upper left')
tikzplotlib.save('../../../Desktop/comparison.tex')
```

```python
with open('./permutated_wilcox_statistics.txt', 'r') as f:
    permutated_wilcox_statistics = f.read().splitlines()
    f.close()
```

```python
observed_wilcox_statistic = 90
```

```python
permutated_wilcox_statistics = [int(float(x)) for x in permutated_wilcox_statistics]
```

```python
plt.hist(
    permutated_wilcox_statistics,
    color='gold',
    label='Permutated U Statistics'
)
plt.vlines(
    x=observed_wilcox_statistic, 
    colors='black',
    label='Observed U Statistic', 
    ymin=0,
    ymax=200
)
plt.title('Permutation Test Using the U statistic between \n the GDP of Countries and their Eigenvector Centralities')
plt.xlabel('U Statistic')
plt.ylabel('Frequency')
plt.legend()
tikzplotlib.save('../../../Desktop/U-statistic.tex')
```
