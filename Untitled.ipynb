```python
import pandas as pd
import seaborn as sns
```

```python
characters_df = pd.read_csv('marvel-characters/characters.csv')
characters2comics_df = pd.read_csv('marvel-characters/charactersToComics.csv')
characters_stats_df = pd.read_csv('marvel-characters/characters_stats.csv')
characters_info_df = pd.read_csv('marvel-characters/marvel_characters_info.csv')
superpowers_df = pd.read_csv('marvel-characters/superheroes_power_matrix.csv')
# other_info_df = pd.read_csv('marvel-characters/marvel_dc_characters.csv')
```

```python
agg_df = pd.merge(
    pd.merge(
        pd.merge(
            characters_df, characters_stats_df, left_on='name', right_on='Name', how='inner'
        ), 
        characters_info_df, on='Name', how='inner'
    ), 
    superpowers_df, on='Name', how='inner'
)
```

```python
agg_df.drop(columns=[
    'name', 'Alignment_y', 'Durability_y', 'Intelligence_y', 'ID'
], axis=1, inplace=True)
agg_df.rename({
    'Alignment_x': 'Alignment', 
    'Durability_x': 'Durability', 
    'Intelligence_x': 'Intelligence'
}, axis=1, inplace=True)
```

```python
agg_df.to_csv('marvel_characters_aggregated.csv', index=False)
```
